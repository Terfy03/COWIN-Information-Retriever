```python
#State information
states = cowin.get_states()
state = [i for i in states["states"] if state_input in i["state_name"].lower()]
```

```python
#Pulling information by district
district_id = str(district[0]['district_id'])
date = date_input
min_age_limit = int(min_age_input)
available_centers_district = cowin.get_availability_by_district(district_id, date, min_age_limit)
```

```python
#Pulling information by pincode
pin_code_input = str(pincode_input)
date = str(date_input)
min_age_limit = int(min_age_input)
available_centers_pincode = cowin.get_availability_by_pincode(pin_code_input, date, min_age_limit)
```

```python
#Function to get vaccine info
def createDf(listInput):
    dfNull = {"date": 'N/A',
              "vaccine": 'N/A',
              "name": 'N/A',
              "address": 'N/A',
              "district_name": 'N/A',
              "block_name": 'N/A',
              "pincode": 'N/A',
              "min_age_limit": 'N/A',
              "fee_type": 'N/A',
              "available_capacity": 'N/A',}
    if len(listInput) == 0:
        return pd.DataFrame(dfNull, index=[0])
    else:
        baseDict = listInput
        info = []
        for i in baseDict:
            relevantData = dict((key,value) for key, value in i.items() if key in ['name','fee_type', "sessions",
                                                                                   'address','block_name','district_name',
                                                                                   'pincode'])
            relevantData["available_capacity"] = relevantData["sessions"][0]["available_capacity"]
            relevantData["min_age_limit"] = relevantData["sessions"][0]["min_age_limit"]
            relevantData["vaccine"] = relevantData["sessions"][0]["vaccine"]
            relevantData["date"] = relevantData["sessions"][0]["date"]
            del relevantData["sessions"]
            info.append(relevantData)

        df = pd.DataFrame(info)
        df_final = df[["date", "vaccine", "name",
                       "address","district_name", "block_name", "pincode",
                       "min_age_limit", "fee_type", "available_capacity"]]
        return pd.DataFrame(df_final)
```

```python
tablePinCode = createDf(available_centers_pincode["centers"])
```
